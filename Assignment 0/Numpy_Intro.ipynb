```python
class DemoObject:
	def __init__(self, x):
		self.x = x
	def get_x_squared(self):
		return (self.x, self.x**2)

for i in range(0, 10):
	demo = DemoObject(i)
	(x, squared) = demo.get_x_squared()
	test = squared > 20
	if test:
		print('x**2 is a big number: ' + str(squared))
	else:
		print('x isn\'t very big yet: ' + str(x))
```

## Numpy Operations

```python
# Simple array creation
a = np.array([0,1,2,3])

# Python type check
print(type(a))

# Numpy type attribute
print(a.dtype)

# Bytes per element
print(a.itemsize)

# Number of dimensions
print(a.ndim)
```

```python
# Array shape - Tuple listing length of array along each dimension
print(a.shape) 
print(np.shape(a))

# Array size - Number of elements in array
print(a.size)
print(np.size(a))

# Memory usage
print(a.nbytes)
```

```python
# Array indexing
print(a[0])
a[0] = 10
print(a)

# Filling an array

# Set all array elements
a.fill(0)
print(a)

# Also works, but may be slower
a[:] = 1
print(a)

# Type coercion
print(a.dtype)

# Assigning float into int array truncates the decimal part
a[0] = 10.6
print(a)

# Same behavior with fill
a.fill(-4.8)
print(a)
```

```python
# Slicing and array

# Slicing arrays
a = np.array([10,11,12,13,14])
print(a[1:3])

# Negative indices count from right
print(a[1:-2])
print(a[-4:3])

# Omitting indices

# Omitted boundaries are assumed to be the beginning (or end)

# Example:

# First three elements
print(a[:3])

# Last two elements
print(a[-2:])

# Every other element
print(a[::2])
```

```python
# Multi-dimensional arrays

# 2D arrays
a = np.array([[ 0, 1, 2, 3], [10,11,12,13]])
print(a)

# Shape = (rows, columns)
print(a.shape)

# Element count
print(a.size)

# Dimension count
print(a.ndim)

# Getting and setting elements
print(a[1,3])
a[1,3] = -1
print(a)

# Access a whole row with one index
print(a[1])

# For columns:
print(a[:,1])

# Slicing:

a = np.array([[ 0, 1, 2, 3, 4, 5],
[10,11,12,13,14,15],
[20,21,22,23,24,25],
[30,31,32,33,34,35],
[40,41,42,43,44,45],
[50,51,52,53,54,55]])

print(a[0,3:5])

print(a[4:,4:])

print(a[:,2])

# Transpose
print(a)
print(a.T)
```

```python
# Array creation

# arange(start, stop=None, step=1, dtype=None)
# Default dtype inferred from start/stop/step datatype
print(np.arange(4))

print(np.arange(0, 2*np.pi, np.pi/4))

# Array of zeros
print(np.zeros((2,4)))

# Array of ones
print(np.ones((3,2),'int64'))

# Identity matrix
a = np.identity(4)
print(a)

# Generate N evenly spaced values between start and stop point
print(np.linspace(0,1,5))

# Generate N evenly spaced values on a log scale between base**start and base**stop
print(np.logspace(0,1,5))
```

### Linear Algebra

```python
# Solving systems of equations

# Columns of a are variable coefficients, rows are individual equations
a = np.array([[-3,1,-1], [-3,0,1],[0,1,-5]])

# Column vector of equation equalities
b = np.array([[-2],[4],[0]])
# Check to ensure matrix is nonsingular
print(np.linalg.det(a))

# Solve
print(np.linalg.solve(a, b))
```

```python
# Matrix multiplication

# * performs elementwise multiplication
a = np.identity(3)
c = np.arange(0,9)
c.shape = (3,3)
print(a*c)

# Use dot for true matrix multiplication
print(a.dot(c))
```

```python
# Eigenvectors and eigenvalues
a = np.array([[-3,1,-1], [-3,0,1],[0,1,-5]])

(val,vec) = np.linalg.eig(a)
print(val)
print(vec)

# Matrix inversion
a_inv = np.linalg.inv(a)
print(a_inv)

# Check to see if this is really the inverse
print(a.dot(a_inv))

# Close enough for floating point
print(np.round(a.dot(a_inv)))
```
